```python
import pandas as pd
import numpy as np
df=pd.read_excel('/Users/ayushjoshi/Downloads/Take Home Assignment Dataset ppython.xlsx')
df.head()
```

```python
df.info()
```

```python
def calculate_group_median(group):
    times = group['Departure Time (Num)'].values
    prices = group['Weighted Average Price'].values
    medians = []
    for t in times:
        mask = np.abs(times - t) <= 2/24
        medians.append(np.median(prices[mask]))
    group['Median Price (Similar Buses)'] = medians
    return group

df = df.groupby('Group Key', group_keys=False).apply(calculate_group_median)
print("Done!")
print(df[['Operator', 'Weighted Average Price', 'Median Price (Similar Buses)']].head(10))
```

```python
# % Difference
df['% Difference'] = (df['Weighted Average Price'] - df['Median Price (Similar Buses)']) / df['Median Price (Similar Buses)']

# Price Flag
def flag(row):
    if row['Operator'] != 'Flixbus':
        return ''
    if row['Median Price (Similar Buses)'] == 0:
        return 'No Comparable Buses'
    elif row['% Difference'] > 0.2:
        return 'Too High'
    elif row['% Difference'] < -0.2:
        return 'Too Low'
    else:
        return 'OK'

df['Price Flag'] = df.apply(flag, axis=1)

# Price Difference
df['Price Difference (Rs)'] = df['Weighted Average Price'] - df['Median Price (Similar Buses)']

print("Done!")
print(df[df['Operator']=='Flixbus'][['Operator','Weighted Average Price','Median Price (Similar Buses)','% Difference','Price Flag','Price Difference (Rs)']].head(10))
```

```python
# Filter only Flixbus rows for output
flixbus_output = df[df['Operator'] == 'Flixbus'][[
    'Route Number', 'Departure Date', 'Operator', 'Bus Type',
    'Is AC', 'Is Seater', 'Is Sleeper', 'Departure Time',
    'Weighted Average Price', 'Median Price (Similar Buses)',
    '% Difference', 'Price Flag', 'Price Difference (Rs)'
]]

# Export to Excel
flixbus_output.to_excel('/Users/ayushjoshi/Downloads/Flixbus_Price_Flagging.xlsx', 
                         sheet_name='Flagging Output', index=False)

print("Done! File saved!")
print(f"Total Flixbus rows: {len(flixbus_output)}")
print(f"Too High: {(flixbus_output['Price Flag']=='Too High').sum()}")
print(f"Too Low: {(flixbus_output['Price Flag']=='Too Low').sum()}")
print(f"OK: {(flixbus_output['Price Flag']=='OK').sum()}")
```
